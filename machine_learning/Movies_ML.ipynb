```python
names=['user_id','item_id','rating','timestamp']
```

```python
movies_dataset =pandas.read_csv(file,delimiter="\t", names=names)
```

```python
# the functions on a dataframe gives a series
listing_series = movies_dataset.groupby('item_id',sort=True).size()
```

```python
#converting series to a dataframe
listing_df = listing_series.to_frame()
```

```python
type(listing_df)
```

```python
listing_df = listing_df.rename(columns={0:"count_times_rated"})
```

```python
listing_df = listing_df.sort_values("count_times_rated", ascending=False)
```

```python
listing_df = listing_df.reset_index()
```

```python
listing_df
```

```python
movies_file="/home/prashant/movilens_dataset/ml-100k/u.item"
```
